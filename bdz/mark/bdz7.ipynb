```python
import numpy as np  # библиотека для матриц и математики
import pandas as pd  # библиотека для работы с табличками
from scipy import stats  # модуль для работы со статистикой

# библиотеки для визуализации
import matplotlib.pyplot as plt
import seaborn as sns

import math

plt.style.use('ggplot')  # стиль графиков

%matplotlib inline
```

```python
xls = pd.ExcelFile(r'./../table.xls')
df = xls.parse(3)
df
```

```python
X = df.C5
Y = df.C4.to_numpy()
```

```python
d = {el: [] for el in X.unique()}
for i in range(len(X)):
    d[X[i]].append(Y[i])
d
```

```python
groups = []
for key in d:
    arr = np.array(d[key])
    groups.append(arr)
    print(f"Выборка \"{key}\"")
    print(f"Объем выборки:", len(d[key]))
    print(f"Cреднее: {round(arr.mean(), 2)}")
    print(f"Дисперсия: {round(arr.var(ddof=1), 2)}")
    print()
```

```python
K = 3
s = 0
n = len(X)
mean = sum(map(sum, groups)) / n

for ar in groups:
    for x in ar:
        s += (x - mean) ** 2
D_x = s / n

#внутригрупповая дисперсиея
s = 0
for ar in groups:
    s = s + len(ar) * np.var(ar)
D_w = s / n

#межгрупповая дисперсияS
s = 0
for ar in groups:
    s = s + len(ar) * (np.mean(ar) - mean) ** 2
D_b = s / n

print(f"{D_b = :.3f}")
print(f'Число степеней свободы: {K - 1}')
print(f'Несмещенная оценка групировочного признака: {(D_b * n) / (K - 1)}')
print()

print(f"{D_w = :.3f}")
print(f'Число степеней свободы: {n - K}')
print(f'Несмещенная оценка остаточного признака {(D_w * n) / (n - K)}')
print()

print(f"{D_x = :.3f}")
print(f'Число степеней свободы: {n - 1}')
print(f'Несмещенная оценка всех признаков: {(D_x * n) / (n - 1)}')
```

```python
math.isclose(D_b + D_w, D_x)
```

```python
eta_squared = D_b / D_x
eta = eta_squared ** (1 / 2)
print(f"Эмпирический коэффициент детерминации: {eta_squared:.3f}")
print(f"Эмпирическое корреляционное отношение: {eta:.3f}")
```
